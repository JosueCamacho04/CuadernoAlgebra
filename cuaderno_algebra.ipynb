## DefiniciÃ³n
Sea $S = \{\textbf{v}_1, \textbf{v}_2, ...., \textbf{v}_n \}$ un conjunto no vacÃ­o de vectores. 

a) Suponga que:

$ \alpha_1 \textbf{v}_1 + \alpha_2 \textbf{v}_2+......+\alpha_2 \textbf{v}_n  = \textbf{0}$ 

implica que $ \alpha_1 = \alpha_2 = ....=\alpha_n$. Se dice entonces que $\textbf{S}$ es ***linealmente independiente***

b) Un conjunto que no es linealmente independiente se llama ***linealmente dependiente***; anÃ¡logamente,  $\textbf{S}$ es linealmente dependiente si y solo si hay escalares $ \alpha_1 = \alpha_2 = ....=\alpha_n$ **NO** todos cero, tales que

## Proceso de Gram-Schmidt

En el espacio euclÃ­deo $â„^3$ con el producto escalar usual definido, se propone un mÃ©todo para encontrar un sistema de vectores, perpendiculares entre sÃ­, a partir de tres vectores no coplanarios cualesquiera. Sean $ð•_1,ð•_2,ð•_3 âˆˆ â„^3$ dichos vectores.
El mÃ©todo consiste de dos proyecciones. La base ortogonal de $â„^3$  compuesta por $ð”_1 = ð•_1$  , se calcula de la siguiente manera.
1.	Se escoge arbitrariamente uno de los vectores dados, por ejemplo,$ð”_1 = ð•_1$.
2.  $ð”_2$ Se calcula como la diferencia entre $ð•_2$  y el vector que resulta de proyectar a $ð•_2$ sobre $ð”_1$ . Dicha diferencia es perpendicular a $ð”_1$  . Es equivalente afirmar que  $ð”_2$  es la diferencia entre $ð•_2$ y el vector que resulta de proyectar a $ð•_2$  sobre la recta que genera $ð”_1$.
3.	$ð”_3$  es la diferencia entre $ð•_3$  y el vector que resulta de proyectar a $ð•_3$  sobre el plano generado por $ð”_1$  y $ð”_2$. La diferencia de vectores tiene como resultado otro vector que es perpendicular al plano.

Esta sencilla interpretaciÃ³n del algoritmo para un caso que puede verse es susceptible de generalizaciÃ³n a espacios vectoriales de dimensiÃ³n arbitraria, con productos internos definidos, no necesariamente canÃ³nicos. Dicha generalizaciÃ³n no es otra que el proceso de Gram-Schmidt.


###Codigo para demostrar ortogonalizacion y ortonormalizacion
```python
import numpy as np

# Definir los vectores
v1 = np.array([[2,-4],[2,1]])
v2 = np.array([[3,2],[-4,-5]])
v3 = np.array([[4,-1,1],[-1,0,4],[-4,-17,-1]])

# Comprobar si los vectores son ortogonales
def are_orthogonal(vectors):
    for i in range(len(vectors)):
        for j in range(i+1, len(vectors)):
            if np.dot(vectors[i], vectors[j]) != 0:
                return False
    return True

if are_orthogonal(v1):
    print("v1 es un conjunto de vectores ortogonales.")
    # Comprobar si los vectores son ortonormales
    if np.linalg.norm(v1[0]) == 1 and np.linalg.norm(v1[1]) == 1:
        print("v1 es un conjunto de vectores ortonormales.")
    else:
        print("v1 no es un conjunto de vectores ortonormales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v1) == 3:
        print("v1 es una base para R3.")
    else:
        print("v1 no es una base para R3.")
else:
    print("v1 no es un conjunto de vectores ortogonales.")

if are_orthogonal(v2):
    print("v2 es un conjunto de vectores ortogonales.")
    # Comprobar si los vectores son ortonormales
    if np.linalg.norm(v2[0]) == 1 and np.linalg.norm(v2[1]) == 1:
        print("v2 es un conjunto de vectores ortonormales.")
    else:
        print("v2 no es un conjunto de vectores ortonormales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v2) == 3:
        print("v2 es una base para R3.")
    else:
        print("v2 no es una base para R3.")

if are_orthogonal(v3):
    print("v3 es un conjunto de vectores ortogonales.")
    # Comprobar si los vectores son ortonormales
    if np.linalg.norm(v3[0]) == 1 and np.linalg.norm(v3[1]) == 1 and np.linalg.norm(v3[2]) == 1:
        print("v3 es un conjunto de vectores ortonormales.")
    else:
        print("v3 no es un conjunto de vectores ortonormales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v3) == 3:
        print("v3 es una base para R3.")
    else:
        print("v3 no es una base para R3.")
else:
    print("v3 no es un conjunto de vectores ortogonales.")

```
